```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
url = "https://en.wikipedia.org/wiki/Capital_punishment_by_country"
html = requests.get(url).text
```

```python
soup = BeautifulSoup(html)
```

```python
tables = soup.find_all("table" ,class_="wikitable")
```

```python
column_names = [item.get_text().strip() for item in tables[0].find_all('th')]
column_names
```

```python
contents = [item.get_text().strip() for item in tables[0].find_all('td')]
contents
```

```python
values = []
for table in tables:
    for item in table.select('td'):
        temp = item.get_text().strip()
        values.append(temp)
values
```

```python
len(values)/6
```

```python
data = np.reshape(values, (236,6))
```

```python
df = pd.DataFrame(data=data, columns=column_names)
df
```

```python
df['Year abolished'] = df['Year abolished'].str.replace('n/a', 'N/A' )
```

```python
df['Year abolished'] = df['Year abolished'].str.contains('N/A')
df
```

```python
df['Year abolished'].value_counts()
```

```python
df = df[df['Year abolished'] == True]
df
```

```python
df.to_csv("project03_capital_punishment_abolished.csv")
```
