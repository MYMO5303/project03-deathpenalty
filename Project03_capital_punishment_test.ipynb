```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/Capital_punishment_by_country"
html = requests.get(url).text
```

```python
soup = BeautifulSoup(html)
```

```python
table = soup.select("table")[0]
print(table.text[:100])
```

```python
header_el = table.select("tr")[0]
```

```python
columns = []
for cell in header_el.select("th"):
    columns.append(cell.text.strip())

columns
```

```python
columns = [ cell.text.strip() for cell in header_el.select("th")]
columns
```

```python
row_els = table.select("tr")[1:]
len(row_els)
```

```python
rows = []

for row in row_els:
    cells = []
    for cell in row.select("td"):
        cells.append(cell.text.strip())
    rows.append(cells)

rows[:2]
```

```python
rows = [
    [ cell.text.strip() for cell in row.select("td") ]
    for row in row_els ]

rows[:2]
```

```python
df = pd.DataFrame(rows, columns=columns)
df.head()
```

```python
df['Notes'] = df['Notes'].str.contains('hunging')
df
```
